```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas_datareader
import datetime
```

```python
import pandas_datareader.data as web
```

```python
start = datetime.datetime(2012,1,1)
end = datetime.datetime(2017,1,1)
tesla = web.DataReader('TSLA', 'yahoo', start, end)
```

```python
tesla.head()
```

```python
ford = web.DataReader("F", "yahoo", start, end)
gm = web.DataReader("GM", "yahoo", start, end)
```

```python
tesla['Open'].plot(label='Tesla',figsize=(15,10),title="Opening Price")
gm['Open'].plot(label='GM')
ford['Open'].plot(label='Ford')
plt.legend();
```

```python
tesla['Volume'].plot(label='Tesla',figsize=(15,10),title="Opening Price")
gm['Volume'].plot(label='GM')
ford['Volume'].plot(label='Ford')
plt.legend();
```

```python
ford['Volume'].max()
```

```python
ford['Volume'].idxmax()
```

```python
ford['Open'].plot(figsize=(20,6))
```

```python
tesla['Total Traded'] = tesla['Open'] * tesla['Volume']
ford['Total Traded'] = ford['Open'] * ford['Volume']
gm['Total Traded'] = gm['Open'] * gm['Volume']
```

```python
tesla['Total Traded'].plot(label='Tesla',figsize=(15,10), title="Open * Vol")
gm['Total Traded'].plot(label='GM')
ford['Total Traded'].plot(label='Ford')
plt.legend();
plt.ylabel('Total Traded');
```

```python
tesla['Total Traded'].idxmax()
```

```python
gm['MA50'] = gm['Open'].rolling(50).mean()
gm['MA200'] = gm['Open'].rolling(200).mean()
gm[['Open','MA50', 'MA200']].plot(label='gm',figsize=(16,8));
```

```python
from pandas.plotting import scatter_matrix
```

```python
car_comp = pd.concat([tesla['Open'],gm['Open'],ford['Open']],axis=1)
```

```python
car_comp.columns = ['Tesla Open','GM Open','Ford Open']
```

```python
car_comp.head()
```

```python
scatter_matrix(car_comp, figsize=(10,10), alpha=0.2, hist_kwds={'bins':50});
```
